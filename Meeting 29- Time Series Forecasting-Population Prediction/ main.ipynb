```python
import pandas as pd

df=pd.read_csv('population.csv')
df.head
```

```python
#filter country that has largest population

highest=df['2020'].nlargest(20).tolist()
filtered_highest=df[df['2020'].isin(highest)]
filtered_highest
```

```python
#visualize compare number of population in 2020

import seaborn as sns
sns.barplot(x='Country Name', y='2020', data=df.tail())
```

```python
#prepare data
#only extract World data - bcs we want to explore only World Population

world=df.loc[df['Country Name']=='World']
world.drop(['Country Name','Country Code','Indicator Name', 'Indicator Code'], axis=1, inplace=True)
world=world.T
world.head
```

```python
 #Label the dataframe to year, population (header column)

world.dropna(inplace=True)
world=world.reset_index().rename(columns={259:'population','index':'year'})
world.tail()
```

```python
#Visualise the development of World Population-scatterplot

sns.lineplot(x='year', y='population', data=world.tail(10))
sns.scatterplot(x='year', y='population', data=world.tail(10))
```

```python
#train Data
#change format dataframe to numeric

world['population']=pd.to_numeric(world['population'])
world ['year']=pd.to_numeric(world['year'])
x=world['year']
y=world['population']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(x,y, test_size=0.2)
X_train=X_train.values.reshape(-1,1)
X_test = X_test.values.reshape(-1, 1)
y_train = y_train.values.reshape(-1, 1)  # Reshape y_train to be 2D
y_test = y_test.values.reshape(-1, 1)   

print(X_train[:5])
print(X_test[:5])
print(y_train[:5])
print(y_test[:5])
```

```python
#Data fitting to train Linear regression

from sklearn.linear_model import LinearRegression
model=LinearRegression()
model=model.fit(X_train, y_train)
y_pred=model.predict(X_test)
y_pred
```

```python
#Compare valuaes y_pred and y_test

import matplotlib.pyplot as plt

plt.xlabel("Year")
plt.ylabel("Population")
plt.title("True vs Predicted")
sns.lineplot(x=X_test.reshape(-1), y=y_pred.reshape(-1))
sns.scatterplot(x=X_test.reshape(-1), y=y_pred.reshape(-1), marker="x", color="green", s=200)
sns.scatterplot(x=X_test.reshape(-1), y=y_test.reshape(-1), color="red")
```

```python
#Predict Futre Population for 2020

test_data=model.predict([[2030]])
test_data
```

```python
#Make prediction based on input provided by user

year_input=int(input("Year:"))
predict_input=model.predict([[year_input]])
print(f"world population prediction in {year_input} is {int(predict_input[0][0])}")
```

```python
#save in pickle

import pickle
pickle.dump(predict_input, open('predmodel.pkl', 'wb'))
```
